# 数据操作

#### 新增列（添加label）

```python
rows_a,cols_a = data_a.shape
rows_b,cols_b = data_b.shape
labels_a = np.zeros(rows_a)
labels_b = np.ones(rows_b)
data_a.insert(0, 'label', labels_a)
data_b.insert(0, 'label', labels_b)
print(data_a.head())
print(data_b.head())
```

#### 修改列数据

```python
data_b['label'] = 0
print(data_b['label'])
data_b['label'] = 1
```

#### 合并数据

```python
data = pd.concat([data_a,data_b], ignore_index = True)
print(data.head())
print(data.tail())
```

#### 数据混序

```python
from sklearn.utils import shuffle
data = shuffle(data)
data.reset_index(drop=True, inplace=True)
print(data['label'][0:10])
```

#### 异常值查看

```python
data['C'] = np.nan
data['D'] = np.nan

data.loc[3:5] = np.nan

data.loc[1,'A'] = np.nan
data.loc[2,'B'] = np.nan

pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
print(data.isna())
```

#### 删除全部为NaN的行
##### NaN = not a number

```python
data.dropna(axis=0,how='all', inplace = True) 
print(data.head())
```

#### 删除全部为NaN的列

```python
data.dropna(axis=1,how='all', inplace = True) 
print(data.head())
```

#### 将NaN值填0

```python
data.fillna(0, inplace = True)
print(data)
```
